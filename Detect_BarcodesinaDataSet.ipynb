```python
!pip install pyzbar
!apt-get install libzbar0

import cv2
import numpy as np
from pyzbar.pyzbar import decode
import os

def detect_barcodes(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Image not found at {image_path}.")
        return 0

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 240)
    cv2.imshow("Edges", edges)

    orb = cv2.ORB_create(nfeatures=500)
    keypoints, descriptors = orb.detectAndCompute(edges, None)
    keypoint_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)

    decoded_objects = decode(keypoint_image)
    barcode_count = 0
    Ncount = 0
    if decoded_objects:
        for obj in decoded_objects:
            barcode_type = obj.type
            barcode_data = obj.data.decode("utf-8")
            print(f"Detected Barcode: Type: {barcode_type}, Data: {barcode_data}")
            barcode_count += 1
    else:
        print("Barcode not detected.")
        Ncount += 1

    return barcode_count, Ncount

def process_images_in_folder(folder_path):
    barcode_total = 0
    N_total = 0
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            print(f"Processing: {filename}")
            barcode_count, Ncount = detect_barcodes(image_path)
            barcode_total += barcode_count
            N_total += Ncount

    print(f"Total barcodes detected: {barcode_total}")
    print(f"Total barcodes not detected: {N_total}")

folder_path = "C:\\Users\\Rajkumar\\OneDrive\\Desktop\\dip_class\\Bar Code Detection.v1i.yolov9\\train\\images"
process_images_in_folder(folder_path)
```

```python
def detect_barcode_from_webcam():
    # Initialize the webcam
    cap = cv2.VideoCapture(0)  # 0 for the default webcam, or change to the appropriate device index

    if not cap.isOpened():
        print("Error: Could not access the webcam.")
        return

    # Get the frames per second (FPS) of the webcam
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Webcam FPS: {fps} frames per second")

    while True:
        # Capture frame-by-frame
        ret, image = cap.read()

        if not ret:
            print("Error: Failed to capture image from webcam.")
            break
```
